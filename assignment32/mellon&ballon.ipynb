```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 800
mellons = pd.DataFrame({'length' : np.random.normal(18.5, 3, N),
                      'width' : np.random.normal(27, 7, N),
                      'weight' : np.random.normal(350,28, N)})

balloons = pd.DataFrame({'length' : np.random.normal(27, 4, N),
                         'width' : np.random.normal(16, 5, N),
                         'weight' : np.random.normal(30, 6, N)})
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
ax.scatter(mellons['width'], mellons['length'], mellons['weight'], c='OrangeRed', marker='^', label='Melons')
ax.scatter(balloons['width'], balloons['length'], balloons['weight'], c='blue', marker='o', label='Balloons')

ax.set_xlabel('Width')
ax.set_ylabel('Length')
ax.set_zlabel('Weight')
ax.legend()

plt.show()
```
